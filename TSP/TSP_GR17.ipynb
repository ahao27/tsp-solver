```python
import numpy as np
import networkx as nx
import random
# matrix = np.loadtxt('map.txt', usecols=range(5),converters=float).astype(np.int64)

def random_map(city=4):
    G = nx.generators.random_graphs.gnp_random_graph(city, 1)    
    for (u,v,w) in G.edges(data=True):
        w['weight'] = random.randint(0,100)
    matrix=nx.to_numpy_array(G)
    return matrix,G


matrix,G=random_map(7)
travel(matrix)
print(matrix)
nx.draw(G)
```
