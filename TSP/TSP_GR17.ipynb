```python
import numpy as np
import networkx as nx
# matrix = np.loadtxt('map.txt', usecols=range(5),converters=float).astype(np.int64)

def random_map(city):
    G = nx.generators.random_graphs.gnp_random_graph(city, 1)
    matrix=nx.adjacency_matrix(G)   
    WeightM=np.random.randint(1,100,size=(7,7))
    WeightM=np.tril(WeightM,k=-1)
    WeightM=WeightM+WeightM.T
    matrix=matrix.toarray()*WeightM
    return matrix,G


matrix,G=random_map(7)
travel(matrix)
print(matrix)
nx.draw(G)
```
